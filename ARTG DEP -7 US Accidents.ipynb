```python
#dropping the rows in the City column which has missing values
df.dropna(subset=["City"],axis = 0, inplace = True)
```

```python
#dropping the rows in the Zipcode column which has missing values
df.dropna(subset=["Zipcode"],axis = 0, inplace = True)
```

```python
#dropping the rows in the Timezone column which has missing values
df.dropna(subset=["Timezone"],axis = 0, inplace = True)
```

```python
#dropping the rows in the Sunrise_Sunset column which has missing values
df.dropna(subset=["Sunrise_Sunset"],axis = 0, inplace = True)
```

```python
#dropping the rows in the Civil_Twilight column which has missing values
df.dropna(subset=["Civil_Twilight"],axis = 0, inplace = True)
```

```python
#dropping the rows in the Nautical and Astronomical Twilight columns which has missing values
df.dropna(subset=["Nautical_Twilight"],axis = 0, inplace = True)
df.dropna(subset=["Astronomical_Twilight"],axis = 0, inplace = True)
```

```python
#Dopping the columns End_Lat and End_lng as they have 99% of missing values
df.drop(['End_Lat','End_Lng'], axis=1)
```

```python
df.drop(['End_Lat','End_Lng'], axis=1, inplace = True)
```

```python
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")  
```

```python
#dropping the Number Column as it has 99% of the rows missing
df.drop(['Number'], axis=1, inplace = True)
```

```python
df.dropna(subset=["Airport_Code"],axis = 0, inplace = True)
```

```python
#To replace the null values in Temperature colummn with the mean value
avg_Temp=df['Temperature(F)'].astype('float').mean(axis=0)
df['Temperature(F)'].replace(np.nan, avg_Temp, inplace=True)
```

```python
#To replace the null values in the Wind_chill column with the corresponsing values in Temperature column as they have high correlation
df['Wind_Chill(F)'].fillna(df['Temperature(F)'], inplace=True)
```

```python
#To replace the null values in Humidity colummn with the mean value
avg_Humidity=df['Humidity(%)'].astype('float').mean(axis=0)
df['Humidity(%)'].replace(np.nan, avg_Humidity, inplace=True)
```

```python
#To replace the null values in Temperature colummn with the mean value
avg_Pressure=df['Pressure(in)'].astype('float').mean(axis=0)
df['Pressure(in)'].replace(np.nan, avg_Pressure, inplace=True)
```

```python
#To replace the null values in Visibility colummn with the mean value
avg_Visibility=df['Visibility(mi)'].astype('float').mean(axis=0)
df['Visibility(mi)'].replace(np.nan, avg_Visibility, inplace=True)
```

```python
#To replace the null values in Wind Direction colummn with the mode value
max_wind=df['Wind_Direction'].value_counts().idxmax()
df["Wind_Direction"].replace(np.nan, max_wind, inplace=True)
```

```python
#To replace the null values in wind speed colummn with the mean value
avg_wind=df['Wind_Speed(mph)'].astype('float').mean(axis=0)
df['Wind_Speed(mph)'].replace(np.nan, avg_wind, inplace=True)
```

```python
#To replace the null values in Weather Condition colummn with the mode value
max_weather=df['Weather_Condition'].value_counts().idxmax()
df["Weather_Condition"].replace(np.nan, max_weather, inplace=True)
```

```python
#To replace the null values in precipitation colummn with the mean value
avg_precip=df['Precipitation(in)'].astype('float').mean(axis=0)
df['Precipitation(in)'].replace(np.nan, avg_precip, inplace=True)
```
